```python
import matplotlib.pyplot as plt
import numpy as np

plt.rcParams['font.sans-serif']=['SimHei'] #用来正常显示中文标签

# 定义全局变量
miu = 0.01 
g = 9.8
m = 70
r = 0.66   # 车轮半径
c = 0.25   # 一个未知系数
w_shoulder = 0.46    # 肩宽
l_leg = 1.0        # 腿长
s = w_shoulder * l_leg   # 人加车正对面积
p = 1.205   # kg/m^3

sigma1 = 42 # 表示有氧呼吸的速率  # sigma2表示无氧呼吸的速率，以函数形式给出
```

```python
# 阻力函数
def f(v):
    return miu*m*g/(2*r) + 0.002*(v)**2  # 1/2*c*p*s   $\approx$  0.002

def f1(v):
    return miu*m*g/(2*r)

def f2(v):
    return 1/2*c*p*s*(v)**2

```

```python
x = np.linspace(0,30,100)
y1=f1(x)
y2=f2(x)
y = y1 + y2


plt.figure(1)
plt.plot(x,y,label='$f_{all}$')
plt.plot([0, 30],[y1, y1],label='$f_{di}$')
plt.plot(x,y2,label='$f_{feng}$')
plt.title("阻力-速度图")
plt.xlabel("$v(m*s^{-1})$")
plt.ylabel("$f(N)$")
plt.legend()
plt.show()

```

```python
# 测试输出
print(20*(f2(20)+y1))

plt.figure(2)
P_zu = y * x
plt.plot(x,P_zu,label='$P_{zu}$')
plt.xlabel("$v(m*s^{-1})$")
plt.ylabel("$P_{zu}(W)$")
plt.legend()
plt.show()

for i in range(10):
    print(i, i*(f2(i)+y1))
```
